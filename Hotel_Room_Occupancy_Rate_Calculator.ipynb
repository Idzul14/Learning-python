```python
# 1. Ask for input (This is where you tell the computer what's happening)
total_rooms = int(input("Enter the total number of rooms in the hotel: "))
occupied_rooms = int(input("Enter the number of rooms currently occupied: "))

# 2. Perform the math
occupancy_rate = (occupied_rooms / total_rooms) * 100

# 3. Display the result
print(f"\n--- Hotel Status Report ---")
print(f"Total Rooms: {total_rooms}")
print(f"Occupied: {occupied_rooms}")
print(f"Occupancy Rate: {occupancy_rate:.1f}%")

# 4. Use an 'if' statement to give feedback
if occupancy_rate >= 80:
    print("Feedback: Excellent! This is peak demand.")
elif occupancy_rate >= 60:
    print("Feedback: Good. This is a solid benchmark for most hotels.")
else:
    print("Feedback: Quiet night. Time to run a promotion?")
```

```python
# 1. Input Section
total_rooms = int(input("Total rooms available in hotel: "))
rooms_sold = int(input("Number of rooms sold tonight: "))
total_revenue = float(input("Total room revenue collected (RM): "))

# 2. Calculations
occupancy_rate = (rooms_sold / total_rooms) * 100
adr = total_revenue / rooms_sold
revpar = total_revenue / total_rooms

# 3. Professional Report Formatting
print("\n" + "="*30)
print("   NIGHTLY REVENUE REPORT")
print("="*30)
print(f"Occupancy Rate: {occupancy_rate:.1f}%")
print(f"ADR:            RM {adr:.2f}")
print(f"RevPAR:         RM {revpar:.2f}")
print("-" * 30)

# 4. Revenue Strategy Analysis
if revpar < (adr * 0.5):
    print("Strategy: Low RevPAR efficiency. Consider 'Flash Sales' to boost occupancy.")
elif adr < 150: # Example benchmark
    print("Strategy: Good occupancy, but ADR is low. Try upselling premium rooms.")
else:
    print("Strategy: Strong performance. Maintain current pricing levels.")
```
