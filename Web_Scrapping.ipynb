```python
# 1. Install the 'library' we need (BeautifulSoup helps us read HTML)
import requests
from bs4 import BeautifulSoup

def get_euro_rate():
    # 2. The URL we want to visit (EUR to MYR)
    url = "https://www.x-rates.com/calculator/?from=EUR&to=MYR&amount=1"

    # 3. Send a 'Request' to the website
    response = requests.get(url)

    # 4. Use BeautifulSoup to parse the HTML (the website's code)
    soup = BeautifulSoup(response.text, 'html.parser')

    # 5. Find the specific 'span' tag that holds the rate
    # On x-rates, the result is usually inside a class called 'ccOutputRslt'
    rate_element = soup.find("span", class_="ccOutputRslt")

    if rate_element:
        rate_text = rate_element.text
        print(f"Current Exchange Rate Found: {rate_text}")

        # Bonus: Let's extract just the number to do math
        rate_number = float(rate_text.split()[0])
        return rate_number
    else:
        print("Could not find the rate. The website structure might have changed!")
        return None

# Run the function
current_rate = get_euro_rate()

if current_rate:
    # 6. Use the rate for a real hospitality scenario
    guest_bill_euro = 170.00
    bill_in_myr = guest_bill_euro * current_rate
    print(f"\nA guest bill of €{guest_bill_euro} equals RM {bill_in_myr:.2f}")
```

```python
import requests
from bs4 import BeautifulSoup

def get_any_rate(from_curr, to_curr):
    # We use 'f-strings' to inject your choices into the URL
    url = f"https://www.x-rates.com/calculator/?from={from_curr}&to={to_curr}&amount=1"

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    try:
        rate_text = soup.find("span", class_="ccOutputRslt").text
        print(f"Update: 1 {from_curr} = {rate_text}")
    except AttributeError:
        print("Error: Could not find that currency pair. Please check the codes (e.g., USD, EUR, MYR).")

# Now you can check anything!
get_any_rate("EUR", "USD")  # Euro to US Dollar
get_any_rate("USD", "MYR")  # US Dollar to Ringgit
get_any_rate("GBP", "EUR")  # British Pound to Euro
```
